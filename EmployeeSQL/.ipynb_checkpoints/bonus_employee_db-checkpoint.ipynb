```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
from sqlalchemy import create_engine
from dotenv import load_dotenv

load_dotenv()
username = os.environ.get('USERNAME')
password = os.environ.get('PASSWORD')
database_uri = 'postgresql://'+username+':'+password+'@localhost:5432/sql_hw'
engine = create_engine(database_uri)
connection = engine.connect()
```

```python
# Create a histogram to visualize the most common salary ranges for employees

# Build the query, and use the result data to construct a data frame 
query1 = "SELECT * FROM salaries"
salary_df = pd.read_sql_query(query1,con=engine)
print(salary_df)
```

```python
# Plot the histogram
salary_df['salary'].hist(bins=15, color='#86bf91',figsize=(12,8))
plt.xlabel("Salary($)", fontsize=12)
plt.ylabel("Employee count", fontsize=12)
plt.title("Salary Ranges of Employees", y=1.02, fontsize=15);
```

```python
# Display the salary ranges and the number of employees in each range
salary_range_bins = pd.cut(salary_df['salary'],bins=10)
labels = salary_range_bins.cat.codes
salary_ranges = salary_range_bins.cat.categories
print(f"Salary Ranges" +"        Employee Count")
print(pd.value_counts(salary_range_bins))

# Plot the histogram
ax = salary_df.hist(column='salary', bins=10, grid=False, figsize=(12,8), color='#86bf91', zorder=2, rwidth=0.9)
plt.xlabel("Salary($)",fontsize=12)
plt.ylabel("Count of Employees",fontsize=12)
plt.title("Salary Ranges of Employees", y=1.02);

# Save Figure
plt.savefig("./Images/salary_range_employees.png")
```

From the histogram plot and the salary range bins and employee counts listed above, we can conclude that 
most common salary ranges for employees are ($43,9910.5-$48,949.2)and ($48,949.2-$57,898.4) 

```python
# Create a bar chart of average salary by title

# Build the query, and use the result data to construct a data frame
query2 = "SELECT titles.title,ROUND(AVG(salaries.salary),2) AS avg_salary FROM titles JOIN salaries ON titles.emp_no = salaries.emp_no GROUP BY titles.title"
avg_salary_title_df = pd.read_sql_query(query2,con=engine)
print(avg_salary_title_df)
```

```python
# Plot the bar chart
sns.set(font_scale=1.4)
ax = avg_salary_title_df.plot.bar(x='title', y='avg_salary', rot=8, color="#86bf91", figsize=(12, 8))
plt.xlabel("Title of the Employee", labelpad=14)
plt.ylabel("Salary($)", labelpad=14)
plt.title("Average Salary by Title", y=1.02);

# Save Figure
plt.savefig("./Images/average_salary_by_title.png")
```
